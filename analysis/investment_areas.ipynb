### Determination of High Need Schools

```python
import pandas as pd

final_data = "../results/final_housing_data.csv"
df = pd.read_csv(final_data)
```

```python
df_ranked = df.sort_values(by="percent_housing_concern", ascending=False, ignore_index=True)
display(df_ranked)
```

```python
# define high need schools as the top 10 schools with highest percentage of students with housing concerns, and at least 22 students with housing concerns
high_need_schools = (
    df.nlargest(10, "percent_housing_concern")
      .query("housing_concern_count >= 22")
)
display(high_need_schools[["school", "housing_concern_count", "total_students", "percent_housing_concern"]])
```
