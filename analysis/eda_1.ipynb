### EDA 1: Summary of Selected Excel Files

```python
import pandas as pd

# load data
kept = pd.read_csv("../results/kept_files.csv")
ignored = pd.read_csv("../results/ignored_files.csv")
mentions = pd.read_csv("../results/housing_mentions.csv")
```

```python
# summary of kept vs ignored files

kept_count = len(kept)
ignored_count = len(ignored)
total_schools = kept_count + ignored_count

summary = pd.DataFrame({
    "category": ["kept", "ignored", "total"],
    "count": [kept_count, ignored_count, total_schools]
})

summary["percent"] = summary["count"] / total_schools * 100

display(summary)
```
